```python

def refined_twist(sauce):
    mid_point = len(sauce) // 2
    return sauce[mid_point:] + sauce[:mid_point]

def refined_bakeMyToss(sauce, light):
    xor_sauce = bytearray()
    for i in range(len(sauce)):
        xor_sauce.append(sauce[i] ^ light[i % len(light)])
    return xor_sauce

original_bytes = b'\x1a\x03\x17\x10\n\x082\x1f\x14?)\n\t62\r]07Z\x11@\x07'
key_phrase = "gdsc is a way of life not just another club"
key_bytes = key_phrase.encode('latin1')

twisted_bytes = refined_twist(original_bytes)
decoded_bytes = refined_bakeMyToss(twisted_bytes, key_bytes)

decoded_message = decoded_bytes.decode('latin1', errors='ignore')
print(decoded_message)
```

```python
import dpkt
import re

def extract_printable_strings(data):
    """Extract readable ASCII strings from binary data."""
    return re.findall(b'[ -~]{4,}', data)

def analyze_pcap(file_path):
    """Analyze a pcap file to find ASCII strings in TCP packet payloads."""
    with open(file_path, 'rb') as f:
        pcap = dpkt.pcap.Reader(f)
        for ts, buf in pcap:
            eth = dpkt.ethernet.Ethernet(buf)
            if not isinstance(eth.data, dpkt.ip.IP):
                continue 

            ip = eth.data
            if isinstance(ip.data, dpkt.tcp.TCP):
                tcp = ip.data
                if tcp.data:
                    strings = extract_printable_strings(tcp.data)
                    for string in strings:
                        try:
                            print(string.decode('ascii'))
                        except UnicodeDecodeError:
                            continue 

if __name__ == '__main__':
    analyze_pcap('./samuel.pcap')
```

```python
import dpkt
import re
import base64

def extract_printable_strings(data):
    """Extract readable ASCII strings from binary data."""
    return re.findall(b'[ -~]{4,}', data)

def decode_base64_strings(strings):
    """Decode Base64 strings and filter for potential flags."""
    flags = []
    for string in strings:
        try:
            decoded = base64.b64decode(string).decode('utf-8')
            if "dsc{" in decoded:  
                flags.append(decoded)
        except (base64.binascii.Error, UnicodeDecodeError):
            continue 
    return flags

def analyze_pcap(file_path):
    """Analyze a pcap file to find ASCII strings and decode Base64 strings in TCP packet payloads."""
    with open(file_path, 'rb') as f:
        pcap = dpkt.pcap.Reader(f)
        potential_base64 = []
        for ts, buf in pcap:
            eth = dpkt.ethernet.Ethernet(buf)
            if not isinstance(eth.data, dpkt.ip.IP):
                continue  

            ip = eth.data
            if isinstance(ip.data, dpkt.tcp.TCP):
                tcp = ip.data
                if tcp.data:
                    strings = extract_printable_strings(tcp.data)
                    for string in strings:
                        if re.fullmatch(b'[A-Za-z0-9+/=]+', string): 
                            potential_base64.append(string)
        
        flags = decode_base64_strings(potential_base64)
        for flag in flags:
            print(f"Possible flag: {flag}")

if __name__ == '__main__':
    pcap_path = 'samuel.pcap'
    analyze_pcap('./samuel.pcap')
```
